```python
import openeo
```

```python
connection = openeo.connect("https://openeo.dataspace.copernicus.eu/openeo/1.2")
```

```python
connection.authenticate_oidc()
```

```python
connection.list_collections()
```

```python
cube = connection.load_collection("SENTINEL_5P_L2",
                                  spatial_extent={"west": 3.75, "east": 4.08, "south": 51.29, "north": 51.39},
                                  temporal_extent=["2021-05-07", "2021-05-14"],
                                  bands=['CO'])
```

```python
connection.
```

```python
bruh = cube.band('CO')
```

```python
a = bruh.raster_to_vector()
```

```python
result = a.save_result('CSV')
```

```python
result
```

```python
a = result.create_job()
```

```python
a.start_and_wait()
a.get_results().download_files("output.csv")
```
